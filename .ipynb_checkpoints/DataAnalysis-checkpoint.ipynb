## House Price Dataset : Data Analysis

In this notebook, we will be analysing different types of variables of the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)

data=pd.read_csv('train.csv')
datafields=pd.read_csv('DataFields.csv')
print('Number of Features : ',data.shape[1])
print('Number of Observations : ',data.shape[0])
data.head()
```

**We will analyse the dataset to identify:**

1. Missing values
2. Numerical variables and their distributions
3. Categorical values, their cardinalities and rare values
4. Outliers
5. Potential relationship between independent variables and target variable (SalePrice)

### Missing values

Let's find out variables with missing values.

```python
var_with_na=[var for var in data.columns 
        if data[var].isnull().sum() > 0
        ]

print('Percentage of missing values\n\n',data[var_with_na].isnull().mean())
```

**Relationship between Sale Price and missing values in each variable**

```python
def vis_with_na(df,var):
    df=df.copy()
    df[var]=np.where(df[var].isnull(),1,0)
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.show()

for var in var_with_na:
    vis_with_na(data,var) 
```

There is difference in average Sale Price of houses when compared between missing information and with information. This information will be taken into account when we engineer these variables.

### Numerical variables

Let's analyse numerical variables in dataset.

```python
var_nums=[column for column in data.columns if data[column].dtypes != 'O']
print('Number of numerical variables :',len(var_nums))
data[var_nums].head()
```

## Temporal variables

There are couple of numerical variables in dataset which have date information. Let's analyse these variables.

```python
var_years=[var for var in data.columns if 'Yr' in var or 'Year' in var]
var_years
```

```python
for var in var_years:
    print(var,data[var].unique())
```

**Let's explore relationship between Sale Price of houses and year variables**

```python
def viz_with_year(var):
    data.groupby(var)['SalePrice'].median().plot()
    plt.title('Change in House Price with '+var)
    plt.ylabel('Median House Price')
    plt.show()
    
for var in var_years:
    viz_with_year(var)    
```

Median house price has increased with later years except for year sold. For year sold, price decreases with years, which is unsual as typically prices go up as years go by.

Let's capture elapsed years between year variables and year in which the house was sold, and try to find relationship if exists. 

```python
def analyse_year_vars(df,var):
    df=df.copy()
    df[var]=df['YrSold']-df[var]
    #df.groupby(var)['SalePrice'].median().plot()
    plt.scatter(df[var],df['SalePrice'])
    plt.xlabel(var)
    plt.ylabel('SalePrice')
    plt.show()

for var in var_years:
     if var != 'YrSold':
         analyse_year_vars(data,var)
```

We see that house price is decreasing in older features. This makes sense because if a house was recently built or remodelled or new garage was built, it typically cost higher.

## Discrete variables

Let's explore discrete variables.

```python
var_disc=[var for var in var_nums 
          if len(data[var].unique())<20 and var not in var_years+['Id']] 

print('Number of discrete variables:',len(var_disc))     
data[var_disc].head()
```

Let's see how house price varies with different values of these discrete variables.

```python
def analyse_discrete(df,var):
    df=df.copy()
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.ylabel('Median SalePrice')
    plt.xlabel(var)
    plt.show()
    
for var in var_disc:
    analyse_discrete(data,var)   
```

We see that there tend to be relationship between Median Sale Price and variables values, however, these relationships are not monotonic, except few like OverallQual. We will make these relationship monotonic for all variables in feature engineering section.

## Continuous variable

Let's analyse continuous variables

```python
var_cont=[var for var in var_nums
          if var not in var_years+var_disc+['Id']]

print('Number of continuous variables :',len(var_cont))
data[var_cont].head()
```

Let's see the distributions

```python
def analyse_continuous(df,var):
    df[var].plot.hist(bins=30)
    plt.title(var)
    plt.xlabel(var)
    plt.ylabel('Number of Houses')
    plt.show()    

for var in var_cont:
    analyse_continuous(data,var)  
```

Variables are not normally distributed. Let's evaluate if a log transform will correct these non-Gaussian distributions.

```python
def analyse_log_continuous(df,var):
    df=df.copy()
    if any(df[var] <= 0):
        pass
    else:
        df[var] = np.log(df[var])
        df[var].hist(bins=30)
        plt.title(var)
        plt.xlabel(var)
        plt.ylabel('Number of Houses')
        plt.show()

for var in var_cont:
    analyse_log_continuous(data,var)    
```

We can apply log tranformation on only 5 variables. But it resulted in better spread of values for these variables.

Let's explore the relationship between these transformed variables and Sale Price.

```python
def analyse_relation_continuous(df,var):
    df=df.copy()
    if any(df[var] <= 0):
        pass
    else:
        plt.scatter(np.log(df[var]),np.log(df['SalePrice']))
        plt.xlabel(var)
        plt.ylabel('SalePrice')
        plt.show()

for var in var_cont:
    if var != 'SalePrice':
        analyse_relation_continuous(data,var)
```

We observe some monotonic association between SalePrice and variables. 

### Outliers

Look out for outliers

```python
for var in var_nums:
    if var in var_years+var_disc+['Id'] or any(data[var] <=0) :
        pass
    else:
        data[var]=np.log(data[var])
        data.boxplot(var)  
        plt.title(var)
        plt.show()     
```

Most of the variables seem to contain outliers. In order to improve performance of a liner model, we should consider removing these outliers.

## Categorical variables

Let's analyse categorical variables in dataset.

```python
var_cat=[var for var in data.columns if data[var].dtype == 'O']        
print('Number of Categorical Variables =',len(var_cat))
data[var_cat].head()
```

**Cardinality**

```python
data[var_cat].nunique()
```

All variables have low cardinality.

**Rare labels**

```python
def rare_labels(df,var):
    temp=df[var].value_counts()/len(df)
    #print(temp)
    print('\n',temp[temp < 0.01])
    
for var in var_cat:
    rare_labels(data,var)   
```

Multiple variables have labels that are present in less than 1% of the houses. We want to remove them in feature engineering section.

Let's explore relationship between Sale Price and Categorical variables

```python
for var in var_cat:
    analyse_discrete(data,var)   
```

We see that categories give information on sale price. We will encode these categorical variables in Feature Engineering section.
